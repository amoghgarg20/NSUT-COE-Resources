# CV Assignment 4

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C://Users/eshub/OneDrive/Desktop/image4.jpg')
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
cv2.imshow??
```

```python
prewittx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewittx_img = cv2.filter2D(src=image, kernel=prewittx, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("Prewitt Image X-axis")
plt.imshow(prewittx_img, cmap='gray')
plt.axis('off')
```

```python
prewitty = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitty_img = cv2.filter2D(src=image, kernel=prewitty, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("Prewitt Image Y-axis")
plt.imshow(prewitty_img, cmap='gray')
plt.axis('off')
```

```python
sobelx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobelx_img = cv2.filter2D(src=image, kernel=sobelx, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("Sobel Image X-axis")
plt.imshow(sobelx_img, cmap='gray')
plt.axis('off')
```

```python
sobely = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
sobely_img = cv2.filter2D(src=image, kernel=sobely, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("Sobel Image Y-axis")
plt.imshow(sobely_img, cmap='gray')
plt.axis('off')
```

```python
logx = np.array([[0, 0, 0], [1, -2, 1], [0, 0, 0]])
logx_img = cv2.filter2D(src=image, kernel=logx, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("LoG Image X-axis")
plt.imshow(logx_img, cmap='gray')
plt.axis('off')
```

```python
logy = np.array([[0, 1, 0], [0, -2, 0], [0, 1, 0]])
logy_img = cv2.filter2D(src=image, kernel=logy, ddepth=0)
plt.figure(figsize=(8,8))
plt.title("LoG Image X-axis")
plt.imshow(logy_img, cmap='gray')
plt.axis('off')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(24,8))

plt.sca(axes[0][0]); 
plt.imshow(prewittx_img); 
plt.title('Prewitt Image X-axis')
plt.axis("off")

plt.sca(axes[0][1]); 
plt.imshow(sobelx_img); 
plt.title('Sobel Image X-axis')
plt.axis("off")

plt.sca(axes[0][2]); 
plt.imshow(logx_img); 
plt.title('LoG Image X-axis')
plt.axis("off")


plt.sca(axes[1][0]); 
plt.imshow(prewitty_img); 
plt.title('Prewitt Image Y-axis')
plt.axis("off")

plt.sca(axes[1][1]); 
plt.imshow(sobely_img); 
plt.title('Sobel Image Y-axis')
plt.axis("off")


plt.sca(axes[1][2]); 
plt.imshow(logy_img); 
plt.title('LoG Image Y-axis')
plt.axis("off")


plt.show()
```
